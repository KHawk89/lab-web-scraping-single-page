```python
from bs4 import BeautifulSoup
```

```python
import requests

r = requests.get('https://www.rollingstone.com/music/music-lists/best-dr-dre-produced-songs-1289410')
r.status_code
```

```python
html = r.content
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
html_table = soup.find_all('article')
html_table
```

```python
titles = []
singers = []


for a in html_table:
    title = a.find('h3', attrs={'class': 'c-title'}).get_text(strip=True) if a.find('h3', attrs={'class': 'c-title'}) else 'Not informed.'
    singer = a.find('span', attrs={'class': 'c-label'}).get_text(strip=True) if a.find('span', attrs={'class': 'c-label'}) else 'Not informed.'
    
    titles.append(title)
    singers.append(singer)
```

```python
singers
```

```python
import pandas as pd
dct = {'singer': singers,'title': titles }

df = pd.DataFrame.from_dict(dct)
df
```
